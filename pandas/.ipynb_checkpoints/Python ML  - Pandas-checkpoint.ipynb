# Selection with loc and iloc

They enable you to select a subset of the rows and columns from a 
DataFrame with NumPy-like notation using either axis labels (loc) or integers
(iloc).

```python
#let’s select a single row and multiple columns by label:
data.loc["Colorado":, ['two', 'three']]
```

```python
#perform some similar selections with integers using iloc:
data.iloc[2, [3, 0, 1]]
data.iloc[2]
data.iloc[[1, 2], [3, 0, 1]]
```
